```python
# questo script analizza il caso b1=b2=inf in cui partiamo da dato iniz. a scalino 
# la X è supposta essere ora diversa da 0, le x sono ancora normali centrate in 0 ma con una loro varianza sigma_t,
# e vediamo a cosa converge a tempi lunghi di ord(1), se è un'altra funzione a gradino e come si sposta

# in un secondo momento le x devono essere prese anche non più centrate in 0, per simulare un passo infinitesimale
# della dinamica di ordine N^2

import numpy as np
import math
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.optimize import fsolve
```

```python
# initialize all the parameters
b1 = 9
b2 = 9
alfa1 = 1
alfa2 = 5
alfa3 = 1
sigma = 1
Nspin = 1000
Niter = 1000000
dt = 0.0001  # così è ordine 1
```

```python
X = 5.5 #initialize the average of the diffusions
sigma = 1

#Xprior = 0

x0 = 0.2
x = np.random.normal(X,sigma,Nspin) 
#print(np.amax(x))
#print(np.amin(x))

########### dati a scalino
m = np.zeros(Nspin)
for i in range(0,Nspin):
    if x[i] > x0:
        m[i] = 1
    else: m[i] = -1
m0 = x0
    
m1 = np.repeat(m[0],Niter) #these will store 
m2 = np.repeat(m[1],Niter) #the first two magnetizations

M = np.repeat(np.mean(m),Niter)  #initialize the average of the magnetizations

print("la media iniziale è:", np.mean(m))

# find the new stable point xo(X), when x0 + X > 0
def f(y):
    return y - 2*norm.cdf(y-X,0,sigma) + X

# find the new stable point xo(X), when x0 + X < 0
def g(y): 
    return y + 2*norm.cdf(-y+X,0,sigma) + X

# equazione per M corretta quando x0(X) > 1-X-M
def h(M):
    return M  - 1 + 2*norm.cdf(1-X-M)

# equazione per M corretta quando x0(X) < -1-X-M 
def l(M):
    return M -1 + 2*norm.cdf(-1-X-M)

print('il valore di x0(X) quando x0(X) + X > 0:', fsolve(f,x0))


#print(np.mean(m))
#print(X)


plt.plot(x,m, 'ro')
plt.xlabel('x')
plt.ylabel('m at final time')
plt.axvline(m0, color='r')
plt.axvline(fsolve(f, x0), color='g')
plt.axvline(fsolve(g,x0), color ='C4')
plt.show()
```

```python
for i in range(0,Niter):
        M[i] = np.mean(m)  #update the running values
        m1[i] = m[0]
        m2[i] = m[1]
        noise = np.random.normal(0,1,Nspin) # draw the updated noise vector 
        u1 = np.random.random_sample(Nspin) # draw N uniform samples to check the update rule 
        u2 = np.random.random_sample(Nspin)
        w1 = Nspin**2*((1 + m)/2)*(1-np.tanh(b1*(m + x) + b2*(M[i] + X)))*dt > u1  #accept the update rules with the right proba
        w2 = Nspin**2*((1 - m)/2)*(1+np.tanh(b1*(m + x) + b2*(M[i] + X)))*dt > u2
        m =  m - w1*(2./Nspin) + w2*(2./Nspin)   
        x = x - (alfa2/Nspin)*(x-X)*dt + sigma/(Nspin**(1/2))*noise*(dt**(1/2))
```

```python
plt.plot(M)
plt.xlabel('time')
plt.ylabel('Media2')
plt.show() 

plt.plot(m1)
plt.xlabel('time')
plt.ylabel('media1')
plt.show() 

plt.plot(m2)
plt.xlabel('time')
plt.ylabel('media1_altra')
plt.show() 

plt.plot(x,m,'ro')
plt.xlabel('x')
plt.axvline(m0, color='r')
plt.ylabel('m at final time')
plt.axvline(fsolve(f, x0), color='g')
plt.axvline(fsolve(g,x0), color ='C4')
plt.ylim(-2,2)
#plt.xlim(-6,0)
plt.show()


#MX = 1- 2*norm.cdf(fsolve(f, x0),Xprior) # + norm.cdf(-fsolve(f, x0),X)
#MGX = 1 - 2*norm.cdf(fsolve(g, x0),Xprior) # + norm.cdf(-fsolve(g, x0),X)

#print('il valore previsto con l eq. per x0(X) quando x0 +X > 0 con Xprior:', MX)
#print('il valore previsto con l eq. per x0(X) quando x0 +X < 0 con Xprior:',MGX)

MX2 = 1- 2*norm.cdf(fsolve(f, x0),X) 
MGX2 = 1 - 2*norm.cdf(fsolve(g, x0),X) 


correct1 = fsolve(h,0.5)
correct2 = fsolve(l,0.5)

print('il valore previsto quando x0(X) > 1-X-M :', correct1)
print('il valore previsto quando x0(X) < -1-X-M :', correct2)


print('il valore previsto con l eq. per x0(X) quando x0 +X > 0 con X:', MX2)
print('il valore previsto con l eq. per x0(X) quando x0 +X < 0 con X:',MGX2)

print('l ultimo valore trovato al tempo finale per la M:', M[Niter-1])

print('il valore di x0(X) quando x0(X) + X > 0:', fsolve(f,x0))
print('il valore di x0(X) quando x0(X) + X < 0:', fsolve(g,x0))

print('per x0(X) + X < 0, check formula x0(X) = -1-X-M:', -1-X-MGX2)
print('per x0(X) + X > 0, check formula x0(X) = 1-X-M:', 1- X - MX2)

# confronta questo punto con l'equazione che abbiamo trovato noi

# e fallo anche per valori iniziali di m non centrati in 0, per studiare la dipendenza 
# dal punto iniziale
```

```python
sigma = 10
X = np.linspace(-5,5, num = 200)

def man_1(x0):
    return x0 + X - 2*norm.cdf(x0-X,0,sigma)

def man_2(x0):
    return x0 + X + 2*norm.cdf(-x0+X,0,sigma)



x0 = fsolve(man_1, np.zeros(200)+1)

x1 = fsolve(man_2, np.zeros(200)+1)


```

```python
# plot delle regioni di piano stabili 
    
plt.plot(x0,X)
plt.plot(x1,X)
plt.plot(-X,X)
plt.xlabel('x0')
plt.ylabel('X')
plt.show() 


# plot della M sui bordi come funzione di X

plt.plot(X, -x0-X+1)
plt.xlabel('X')
plt.ylabel('M sul bordo dx')
plt.show()


plt.plot(X, -x1-X-1)
plt.xlabel('X')
plt.ylabel('M sul bordo sx')
plt.show()
```
