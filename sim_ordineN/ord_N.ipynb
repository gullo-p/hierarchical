```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
# initialize all the parameters
b1 = 0.5
b2 = 5
alfa1 = 1
alfa2 = 1
alfa3 = 1
sigma = 0.1
Nspin = 1000
Niter = 10000000
dt = 0.0005 #questo Ã¨ ordine N
```

```python
#X = np.random.normal(0,1,Nspin)  #the vector of diffusions
X = np.repeat(0.,Nspin)

MediaX = np.repeat(np.mean(X),Niter)  #initialize the average of the diffusions

eps = 1  #if you want to put a bias in the initial distribution 

x1 = np.repeat(0.,Niter) #this will store the first diffusion

# m = np.repeat(1.,Niter)

#m = -1+eps + (2-eps)*np.random.random_sample(Nspin) #initialize the magnetizations vector
p = 2/3 # proportion of spin one's
m = np.zeros(Nspin) + 2*p - 1

m1 = np.repeat(m[0],Niter) #these will store 
m2 = np.repeat(m[1],Niter) #the first two magnetizations

M = np.repeat(np.mean(m),Niter)  #initialize the average of the magnetizations
```

```python
for i in range(0,Niter):
        M[i] = np.mean(m)  #update the running values
        m1[i] = m[0]
        m2[i] = m[1]
        x1[i] = X[0]
        MediaX[i] = np.mean(X)
        u1 = np.random.random_sample(Nspin) # draw N uniform samples to check the update rule 
        u2 = np.random.random_sample(Nspin)
        noise = np.random.normal(0,1,Nspin) # draw the updated noise vector 
        w1 = Nspin*(1 + m)/2*np.exp(-b1*m - b2*M[i] - b1*X)*dt > u1  #accept the update rules with the right proba
        w2 = Nspin*(1 - m)/2*np.exp(b1*m + b2*M[i] + b1*X)*dt > u2
        X = X - (alfa2/Nspin)*X*dt + sigma/(Nspin**(1/2))*noise*(dt**(1/2))
        m =  m - w1*(2/Nspin) + w2*(2/Nspin)   
```

```python
plt.plot(M)
plt.xlabel('time')
plt.ylabel('Media2')
plt.show() 

plt.plot(m1)
plt.xlabel('time')
plt.ylabel('media1')
plt.show() 

plt.plot(m2)
plt.xlabel('time')
plt.ylabel('media1_altra')
plt.show() 

plt.plot(MediaX)
plt.xlabel('time')
plt.ylabel('MediaX')
plt.show() 

plt.plot(x1)
plt.xlabel('time')
plt.ylabel('prima diffusione')
plt.show() 
```
